## 4. Data Preprocessing <a id='preprocess'></a>

```python
# Remove the features that we won't use
df.drop(['title', 'date_posted'], axis=1, inplace=True)
df.info()
```

```python
# Lets drop the NULL values
df = df.dropna().reset_index(drop=True)
df.describe(include='all').transpose()
```

```python
# Lets drop if there are any duplicate values
df = df.drop_duplicates().reset_index(drop=True)
df.describe(include='all').transpose()
```

```python
# First the data type of the price should be corrected
df['price_in_aed'] = df['price_in_aed'].str.replace(',', '').astype(int)
df.describe(include='all').transpose()
```

```python
# Plotting the price to see the distribution and to check if there are any outliers.
fig = plt.figure(figsize=(14,6))

ax1 = fig.add_subplot(1,2,1)
#ax1 = sb.distplot(df['price_in_aed'])
ax1 = sb.histplot(data=df, x='price_in_aed', kde=True)
ax1.set_title('Price Distribution', fontsize=18)
ax1.ticklabel_format(style='plain', axis='x') # To repress scientific notation

ax2 = fig.add_subplot(1,2,2)
ax2 = sb.boxplot(df['price_in_aed'])

plt.show()
```

üîé **Observation:** As can be seen on the above graph the price above 200,000 AED should be removed from the dataset.

```python
df = df[df['price_in_aed'] <= 200000]
```

```python
# Plotting the price to see the distribution after removing the outliers
fig = plt.figure(figsize=(14,6))

ax1 = fig.add_subplot(1,2,1)
ax1 = sb.histplot(data=df, x='price_in_aed', kde=True)
ax1.set_title('Price Distribution', fontsize=18)
ax1.ticklabel_format(style='plain', axis='x') # To repress scientific notation

ax2 = fig.add_subplot(1,2,2)
ax2 = sb.boxplot(df['price_in_aed'])

plt.show()
```

### Handling Categorical Features

The performance of a machine learning also on how we process and feed different types of variables to the model. The number of different values in `motors_trim`, `model` and `company` are too high. Considering other categorical features, this might impact the performance of the model. That's why it's better to handle these feautes other than one hot encoding or dummy encoding. Considering that these features have high relation with `price_in_aed` it's better to apply target encoding. But before performing any type of encoding, it's better to clean the dataset from categories with very few enteries, which might lead to overfitting.

```python
df.groupby('company')['company'].count().sort_values()
```

```python
# Filter companies which has less than 20 enteries
a = df.groupby('company')['company'].count()
a = a.where(a < 20).dropna()
a.index
```

```python
companies_to_filter = a.index
df = df[~df['company'].isin(companies_to_filter)]
df.shape
```

```python
df.groupby('model')['model'].count().sort_values()
```

```python
# Filter moldes which has less than 20 enteries
a = df.groupby('model')['model'].count()
a = a.where(a < 20).dropna()
a.index

models_to_filter = a.index
df = df[~df['model'].isin(models_to_filter)]
df.shape
```
